Task 6. Adding HDI Category Column

```python
def classify_hdi(hdi):
    if hdi < 0.550:
        return "Low"
    elif 0.550 <= hdi <= 0.699:
        return "Medium"
    elif 0.700 <= hdi <= 0.799:
        return "High"
    else:
        return "Very High"
#creating new HDI category column
hdi_2022_df['hdi_category'] = hdi_2022_df['hdi'].apply(classify_hdi)
print(hdi_2022_df['hdi_category'].value_counts())
#checking rows
hdi_2022_df[['country', 'hdi', 'hdi_category']].head(10)
#saving final dataframe as csv
hdi_2022_df.to_csv("hdi_category_added.csv", index=False)
print("File saved as: hdi_category_added.csv")
```

Task 3. Outlier Detection

```python
south_asia_2022 = south_asia_data[south_asia_data['year'] == 2022].copy()

# Calculate Q1, Q3, and IQR for HDI
Q1_hdi = south_asia_2022['hdi'].quantile(0.25)
Q3_hdi = south_asia_2022['hdi'].quantile(0.75)
IQR_hdi = Q3_hdi - Q1_hdi

hdi_lower = Q1_hdi - 1.5 * IQR_hdi
hdi_upper = Q3_hdi + 1.5 * IQR_hdi

# Calculate Q1, Q3, and IQR for GNI per Capita
Q1_gni = south_asia_2022['gross_inc_percap'].quantile(0.25)
Q3_gni = south_asia_2022['gross_inc_percap'].quantile(0.75)
IQR_gni = Q3_gni - Q1_gni

gni_lower = Q1_gni - 1.5 * IQR_gni
gni_upper = Q3_gni + 1.5 * IQR_gni

# Identify outliers
outliers = south_asia_2022[(south_asia_2022['hdi'] < hdi_lower) | (south_asia_2022['hdi'] > hdi_upper) |(south_asia_2022['gross_inc_percap'] < gni_lower) | (south_asia_2022['gross_inc_percap'] > gni_upper)]

# Create scatter plot with highlighted outliers
plt.figure(figsize=(10, 6))
# Normal points
normal = south_asia_2022[~south_asia_2022.index.isin(outliers.index)]
plt.scatter(normal['gross_inc_percap'], normal['hdi'], color='blue', s=80, alpha=0.6, label='Normal Countries')
# Outliers
plt.scatter(outliers['gross_inc_percap'], outliers['hdi'],  color='red', s=150, alpha=0.8, label='Outliers', edgecolors='black', linewidths=2)

plt.xlabel('GNI per Capita ($)')
plt.ylabel('HDI')
plt.title('Outlier Detection: GNI per Capita vs HDI (South Asia 2022)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

Task 4. Metric Comparisons

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Load datasets (assuming you have already saved them in Step 1)
south_asia = pd.read_csv("HDI_SouthAsia_2020_2022.csv")
middle_east = pd.read_csv("HDI_MiddleEast_2020_2022.csv")

# Metrics to compare
metrics = ['hdi', 'gender_development', 'life_expectancy', 'gross_inc_percap']

# Compute mean values for each metric in both regions
mean_SA = south_asia[metrics].mean()
mean_ME = middle_east[metrics].mean()

# Compute absolute disparity to highlight the largest difference
disparity = abs(mean_SA - mean_ME)
max_idx = np.argmax(disparity)

# X positions for the metrics
x = np.arange(len(metrics))
width = 0.35  # width of each bar

# Plot grouped bar chart
plt.figure(figsize=(10,6))

# South Asia bars
colors_SA = ['skyblue']*len(metrics)
colors_SA[max_idx] = 'dodgerblue'  # highlight greatest disparity
plt.bar(x - width/2, mean_SA, width, label='South Asia', color=colors_SA)

# Middle East bars
colors_ME = ['lightgreen']*len(metrics)
colors_ME[max_idx] = 'limegreen'  # highlight greatest disparity
plt.bar(x + width/2, mean_ME, width, label='Middle East', color=colors_ME)

# Labels and title
metric_labels = ['HDI', 'Gender Dev Index', 'Life Expectancy', 'Gross_inc_per Capita']
plt.xticks(x, metric_labels, rotation=15)
plt.ylabel('Mean Value')
plt.xlabel('Metric')
plt.title('Metric Comparison: South Asia vs Middle East (2020–2022)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)  # horizontal gridlines

# Optional: annotate mean values on top of bars
for i in range(len(metrics)):
    plt.text(x[i] - width/2, mean_SA[i]+0.01, f"{mean_SA[i]:.2f}", ha='center', va='bottom', fontsize=9)
    plt.text(x[i] + width/2, mean_ME[i]+0.01, f"{mean_ME[i]:.2f}", ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

# Print which metric shows the greatest disparity
print(f"The metric with the greatest disparity between regions is '{metric_labels[max_idx]}'.")
```

Task 7. Outlier Detection

```python
import pandas as pd
import matplotlib.pyplot as plt

# HDI Outlier calculation for South Asia
Q1_hdi = south_asia['hdi'].quantile(0.25)
Q3_hdi = south_asia['hdi'].quantile(0.75)
IQR_hdi = Q3_hdi - Q1_hdi
hdi_lower = Q1_hdi - 1.5 * IQR_hdi
hdi_upper = Q3_hdi + 1.5 * IQR_hdi

# GNI Outlier calculation for South Asia
Q1_gni = south_asia['gross_inc_percap'].quantile(0.25)
Q3_gni = south_asia['gross_inc_percap'].quantile(0.75)
IQR_gni = Q3_gni - Q1_gni
gni_lower = Q1_gni - 1.5 * IQR_gni
gni_upper = Q3_gni + 1.5 * IQR_gni

# Detect outliers
outliers_SA = south_asia[
    (south_asia['hdi'] < hdi_lower) | (south_asia['hdi'] > hdi_upper) |
    (south_asia['gross_inc_percap'] < gni_lower) | (south_asia['gross_inc_percap'] > gni_upper)
]

print("South Asia Outliers:")
print(outliers_SA[['country', 'hdi', 'gross_inc_percap']])

# Scatter plot
plt.figure(figsize=(8,6))
plt.scatter(south_asia['gross_inc_percap'], south_asia['hdi'], label='Normal Countries')
plt.scatter(outliers_SA['gross_inc_percap'], outliers_SA['hdi'], color='red', label='Outliers')
plt.xlabel('GNI per Capita')
plt.ylabel('HDI')
plt.title('South Asia: Outlier Detection (HDI vs GNI per Capita)')
plt.legend()
plt.show()
```
